```python
import pandas as pd
```

```python
delivery_data = pd.read_csv('deliveries.csv')
delivery_data.head()
```

```python
#Finding winning and losing team
```

```python
team1_chasing = match_data.query("team2 == toss_winner and toss_decision == 'bat' and team1 == winner")
```

```python
team2_chasing = match_data.query("team1 == toss_winner and toss_decision == 'bat' and team2 == winner")
```

```python
chasing = pd.concat([team1_chasing, team2_chasing])
```

```python
##Finding which toss decision is best for each ground
```

```python
match_data['venue'].unique()
```

```python
for value in match_data['venue'].unique().tolist():
    venues = match_data.groupby('venue')
    venue_data = venues.get_group(value)
    fielding_count = venue_data.loc[(venue_data['toss_winner'] == venue_data['winner']) & (venue_data['toss_decision'] == 'field')]
    batting_count = venue_data.loc[(venue_data['toss_winner'] == venue_data['winner']) & (venue_data['toss_decision'] == 'bat')]  
    fielding = fielding_count['winner'].value_counts()
    batting = batting_count['winner'].value_counts()
    if fielding.sum() > batting.sum():
        print(f'Venue: {value}, with toss decision as FIELDING with fielding win count = {fielding.sum()}')
    else:
        print(f'Venue {value}, with toss decision as BATTING with batting win count = {batting.sum()}')
```
