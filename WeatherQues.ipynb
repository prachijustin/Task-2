```python
new_data = pd.read_csv('testset.csv', parse_dates=['datetime_utc'])
```

```python
new_data
```

```python
type(new_data.datetime_utc[0])
```

```python
#Now since datetime_utc is unique for each row, we set it to index value
```

```python
new_data.set_index('datetime_utc', inplace=True)
```

```python
new_data.head()
```

```python
#Here datetime_utc values are continuous for each hour, each day.
#But at some places, it's missing. 
#For eg. datetime_utc for 1996-11-01 15:00:00 is missing between 1996-11-01 14:00:00 and 1996-11-01 16:00:00
#Similary, there are other fields like this as well.
#So we distribute data equally within hours
```

```python
dt = pd.date_range('1996-11-01 11:00:00', '2017-04-24 18:00:00', freq='60min')
```

```python
idx = pd.DatetimeIndex(dt)
```

```python
new_data = new_data.reindex(idx)
new_data
```

```python
#Dropping column with all NaNs
#Since a column with all value set as NaN is of no use(We cannot get any data from that column)
#We will drop that column
```

```python
new_data = new_data.dropna(axis = 1, how = 'all')
new_data.head()
```

```python
#Here _precipm column has all values as NaN. So it has been dropped now
```

```python
new_data = new_data.replace(-9999.0, np.NaN)
```

```python
new_data.head()
```

```python
new_data.fillna(method='ffill', inplace=True)
```

```python
new_data
```

```python
new_data = new_data.fillna(0)
```

```python
new_data
```

```python
new_data = new_data.interpolate(method='time')
```

```python
new_data
```

```python
new_data.loc[new_data[' _conds'] == 'Unknown']
```

```python
new_data = new_data.replace('Unknown',np.nan).ffill()
```

```python
year_data = new_data.groupby(new_data.index.year).mean()
```

```python
def create_plot(ptype):
    # setting the x-axis vaues
    x = new_data.index.year.unique()
     
    if ptype == 'dewptm':
        y = year_data[' _dewptm'].tolist()
    elif ptype == 'fog':
        y = year_data[' _fog'].tolist()
    elif ptype == 'hail':
        y = year_data[' _hail'].tolist()
    elif ptype == 'heatindexm':
        y = year_data[' _heatindexm'].tolist()
    elif ptype == 'hum':
        y = year_data[' _hum'].tolist()
    elif ptype == 'pressurem':
        y = year_data[' _pressurem'].tolist()
    elif ptype == 'rain':
        y = year_data[' _rain'].tolist()
    elif ptype == 'snow':
        y = year_data[' _snow'].tolist()
    elif ptype == 'tempm':
        y = year_data[' _tempm'].tolist()
    elif ptype == 'thunder':
        y = year_data[' _thunder'].tolist()
    elif ptype == 'tornado':
        y = year_data[' _tornado'].tolist()
    elif ptype == 'vism':
        y = year_data[' _vism'].tolist()
    elif ptype == 'wdird':
        y = year_data[' _wdird'].tolist()
    elif ptype == 'wgustm':
        y = year_data[' _wgustm'].tolist()
    elif ptype == 'windchillm':
        y = year_data[' _windchillm'].tolist()
    elif ptype == 'wspdm':
        y = year_data[' _wspdm'].tolist()
             
    return(x, y)
```

```python
fig = plt.figure()

plt1 = fig.add_subplot(221)
plt2 = fig.add_subplot(222)
plt3 = fig.add_subplot(223)
plt4 = fig.add_subplot(224)

x, y = create_plot('dewptm')
plt1.plot(x,y)
plt1.set_title('Dewptm')

x, y = create_plot('fog')
plt2.plot(x,y)
plt2.set_title('Fog')

x, y = create_plot('hail')
plt3.plot(x,y)
plt3.set_title('Hail')

x, y = create_plot('heatindexm')
plt4.plot(x,y)
plt4.set_title('Heatindexm')

fig.subplots_adjust(hspace=0.5,wspace=0.5)

plt.show()
```
